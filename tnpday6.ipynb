## Linear LL

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insertLeft(self, e):
        n = Node(e)
        if self.head is None:
            self.head = n
        else:
            n.next = self.head
            self.head = n
        print(e, "inserted to the left of the head")

    def delLeft(self):
        if self.head is None:
            print("Empty")
        else:
            t = self.head
            self.head = self.head.next
            print(t.data, "Left node deleted")

    def insertRight(self, e):
        n = Node(e)
        if self.head is None:
            self.head = n
        else:
            t = self.head
            while t.next is not None:
                t = t.next
            t.next = n
        print(e, "inserted to the right of the head")

    def delRight(self):
        if self.head is None:
            print("Empty")
        else:
            t2 = t = self.head
            while t.next is not None:
                t2 = t
                t = t.next
            if t == self.head:
                self.head = None
            else:
                t2.next = None
            print(t.data, "Right node deleted")

    def display(self):
        if self.head is None:
            print("Empty")
        else:
            t = self.head
            while t is not None:
                print(t.data, end=" -> ")
                t = t.next
        print()
    def dhund(self,e):
        if self.head is None:
            print("Empty")
        else:
            count=0
            t = self.head
            while t is not None:
                count+=1
                if t.data==e:
                    flag=1
                    break
                t = t.next
            if flag==1:
                print("Element found at position",count)
            else:
                print("Element not found")
        print()

l = LinkedList()
print("Linear Linked List")
print("1. Insert Left, 2. Delete Left, 3. Insert Right, 4. Delete Right, 5. Display, 6. Exit")
while True:
    ch = int(input("Enter your choice: "))
    if ch == 1:
        e = int(input("Enter element: "))
        l.insertLeft(e)
    elif ch == 2:
        l.delLeft()
    elif ch == 3:
        e = int(input("Enter element: "))
        l.insertRight(e)
    elif ch == 4:
        l.delRight()
    elif ch == 5:
        l.display()
    elif ch == 6:
        e=int(input("Enter element to be searched: "))
        l.dhund(e)
        break
    else:
        print("Invalid Choice")
````

## Circular LL

> has  root & last

Applications:
+ Circular Buffer
+ playlist/songs
+ data filteration 

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None
        self.last = None

    def insertLeft(self, e):
        n = Node(e)
        if self.head is None:
            self.head = n
            self.last = n
            self.last.next = self.head
        else:
            n.next = self.head
            self.head = n
            self.last.next = self.head
        print(e, "inserted to the left of the head")

    def delLeft(self):
        if self.head is None:
            print("Empty")
        else:
            t = self.head
            if self.head == self.last:
                self.head = None
                self.last = None
            else:
                self.head = self.head.next
                self.last.next = self.head
            print(t.data, "Left node deleted")

    def insertRight(self, e):
        n = Node(e)
        if self.head is None:
            self.head = n
            self.last = n
            self.last.next = self.head
        else:
            self.last.next = n
            self.last = n
            self.last.next = self.head
        print(e, "inserted to the right of the head")

    def delRight(self):
        if self.head is None:
            print("Empty")
        else:
            t2 = t = self.head
            if self.head == self.last:
                self.head = None
                self.last = None
            else:
                while t.next != self.head:
                    t2 = t
                    t = t.next
                self.last = t2
                self.last.next = self.head
            print(t.data, "Right node deleted")

    def display(self):
        if self.head is None:
            print("Empty")
        else:
            t = self.head
            while True:
                print(t.data, end=" -> ")
                if t == self.last:
                    break
                t = t.next
        print()

    def dhund(self, e):
        if self.head is None:
            print("Empty")
        else:
            count = 1  
            t = self.head
            flag = 0  
            while True:
                if t.data == e:
                    flag = 1
                    break
                if t == self.last:
                    break
                t = t.next
                count += 1  
            if flag == 1:
                print("Element found at position", count)
            else:
                print("Element not found")
        print()

l = CircularLinkedList()
print("Circular Linked List")
print("1. Insert Left, 2. Delete Left, 3. Insert Right, 4. Delete Right, 5. Display, 6. Search, 7. Exit")
while True:
    ch = int(input("Enter your choice: "))
    if ch == 1:
        e = int(input("Enter element: "))
        l.insertLeft(e)
    elif ch == 2:
        l.delLeft()
    elif ch == 3:
        e = int(input("Enter element: "))
        l.insertRight(e)
    elif ch == 4:
        l.delRight()
    elif ch == 5:
        l.display()
    elif ch == 6:
        e = int(input("Enter element to be searched: "))
        l.dhund(e)
    elif ch == 7:
        break
    else:
        print("Invalid Choice")
````

## Data structures using LL



> STACK
+ insert left for push
+ Del left for pop
+ print for display

since LL is a dynamic ds there is no need of isfull() or size
this will create a dynamic stack


````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class stackll:
    def __init__(self):
        self.tos = None

    def push(self, e):
        n = Node(e)
        if self.tos is None:
            self.tos = n
        else:
            n.next = self.tos
            self.tos = n
        print(e, "pushed to the stack")
    
    def pop(self):
        if self.tos is None:
            print("Empty")
        else:
            t = self.tos
            self.tos = self.tos.next
            print(t.data, "popped from the stack")
    
    def display(self):
        if self.tos is None:
            print("Empty")
        else:
            t = self.tos
            while t is not None:
                if t.next is None:
                    print(t.data, end=" ")
                else:
                    print(t.data, end=" -> ")
                t = t.next
        print()

    def peek(self):
        if self.tos is None:
            print("Empty")
        else:
            print(self.tos.data, "is at the top of the stack")

s = stackll()
print("Stack using Linked List")
print("1. Push, 2. Pop, 3. Display, 4. Peek, 5. Exit")
while True:
    ch = int(input("Enter your choice: "))
    if ch == 1:
        e = int(input("Enter element: "))
        s.push(e)
    elif ch == 2:
        s.pop()
    elif ch == 3:
        s.display()
    elif ch == 4:
        s.peek()
    elif ch == 5:
        break
    else:
        print("Invalid Choice")
````

> Queue
+ insert right for enqueue
+ delete left for dequeue
+ dynamic queue

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class queue:
    def __init__(self):
        self.front = None
        
    def enqueue(self, e):
        n = Node(e)
        if self.front is None:
            self.front = n
        else:
            t = self.front
            while t.next is not None:
                t = t.next
            t.next = n
        print(e, "enqueued to the queue")

    def dequeue(self):
        if self.front is None:
            print("Empty")
        else:
            t = self.front
            self.front = self.front.next
            print(t.data, "dequeued from the queue")

    def display(self):
        if self.front is None:
            print("Empty")
        else:
            t = self.front
            while t is not None:
                if t.next is None:
                    print(t.data, end=" ")
                else:
                    print(t.data, end=" -> ")
                t = t.next
        print()

    def peek(self):
        if self.front is None:
            print("Empty")
        else:
            print(self.front.data, "is at the front of the queue")

q = queue()
print("Queue using Linked List")
print("1. Enqueue, 2. Dequeue, 3. Display, 4. Peek, 5. Exit")
while True:
    ch = int(input("Enter your choice: "))
    if ch == 1:
        e = int(input("Enter element: "))
        q.enqueue(e)
    elif ch == 2:
        q.dequeue()
    elif ch == 3:
        q.display()
    elif ch == 4:
        q.peek()
    elif ch == 5:
        break
    else:
        print("Invalid Choice")
````

## Doubly linked list
+ Supports bidirectional traversal (forward and backward).
+ Efficient for inserting and deleting elements at both ends.
+ Requires more memory due to the additional "prev" pointers.
+ Useful for applications requiring bidirectional navigation.
+ Can be complex to implement due to managing "prev" pointers.
+ Foundation for more complex structures like deques.
+ Circular doubly linked lists form loops and are used in cyclic scenarios.

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insertLeft(self, e):
        n = Node(e)
        if self.head is None:
            self.head = n
        else:
            n.next = self.head
            self.head.prev = n
            self.head = n
        print(e, "inserted to the left of the head")

    def delLeft(self):
        if self.head is None:
            print("Empty")
        else:
            t = self.head
            self.head = self.head.next
            if self.head is not None:
                self.head.prev = None
            print(t.data, "Left node deleted")

    def insertRight(self, e):
        n = Node(e)
        if self.head is None:
            self.head = n
        else:
            t = self.head
            while t.next is not None:
                t = t.next
            t.next = n
            n.prev = t
        print(e, "inserted to the right of the head")

    def delRight(self):
        if self.head is None:
            print("Empty")
        else:
            t2 = t = self.head
            while t.next is not None:
                t2 = t
                t = t.next
            if t == self.head:
                self.head = None
            else:
                t2.next = None
            print(t.data, "Right node deleted")

    def display(self):
        if self.head is None:
            print("Empty")
        else:
            t = self.head
            while t is not None:
                if t.next is None:
                    print(t.data, end=" ")
                else:
                    print(t.data, end=" <-> ")
                t = t.next
        print()

    def disprev(self):
        if self.head is None:
            print("Empty")
        else:
            t = self.head
            while t is not None:
                t=t.next
            while t is not None:
                if t.prev is None:
                    print(t.data, end=" ")
                else:
                    print(t.data, end=" <-> ")
                t = t.prev
        print()

    def dhund(self, e):
        if self.head is None:
            print("Empty")
        else:
            count = 1
            t = self.head
            flag = 0
            while t is not None:
                if t.data == e:
                    flag = 1
                    break
                t = t.next
                count += 1
            if flag == 1:
                print("Element found at position", count)
            else:
                print("Element not found")
        print()

l = DoublyLinkedList()
print("Doubly Linked List")
print("1. Insert Left, 2. Delete Left, 3. Insert Right, 4. Delete Right, 5. Display, 6. Search, 7. Exit")
while True:
    ch = int(input("Enter your choice: "))
    if ch == 1:
        e = int(input("Enter element: "))
        l.insertLeft(e)
    elif ch == 2:
        l.delLeft()
    elif ch == 3:
        e = int(input("Enter element: "))
        l.insertRight(e)
    elif ch == 4:
        l.delRight()
    elif ch == 5:
        l.display()
    elif ch == 6:
        e = int(input("Enter element to be searched: "))
        l.dhund(e)
    elif ch == 7:
        break
    else:
        print("Invalid Choice")
````

## Threaded Binary Tree

A threaded binary tree is a binary tree data structure in which every node maintains additional references or threads that allow for efficient in-order traversal without using recursive methods or a stack for navigation. In a threaded binary tree, there are two types of threads:

In-order Thread: For each node, there is a thread that points to the in-order predecessor or successor of that node. This allows for easy navigation without the need for recursive function calls.

Null Thread: A null thread indicates that a child of a node is empty, which simplifies checking for empty children during traversal.

Threaded binary trees are particularly useful for situations where in-order traversals are frequent and efficiency is critical. They save memory compared to traditional recursive approaches, which rely on a call stack for traversal.

One-Handed and Two-Handed Trees:

The terms "one-handed" and "two-handed" are typically used in the context of threaded binary trees:

One-Handed Trees: A one-handed tree, also known as a single-threaded tree, has either in-order predecessor threads or in-order successor threads for efficient in-order traversal. It's essentially a threaded binary tree with threads in only one direction (either predecessor threads or successor threads). One-handed trees are simpler to implement and require less memory compared to two-handed trees.

Two-Handed Trees: A two-handed tree, also known as a double-threaded tree, has both in-order predecessor threads and in-order successor threads. This design allows for bidirectional in-order traversal and is a more advanced implementation compared to one-handed trees. Two-handed trees are used in situations where traversal in both directions is required and are less common than one-handed trees.
